```python
import re
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

## Performing bag-of-words using plan python

```python
# Sample text
data = '''In the past fifteen years, the grow of using Bag of Words (BoW) method in the field of computer 
vision is visibly observed. In addition, for the text classification and texture recognition, it can also 
be used in classification of images, videos, robot localization, etc. It is one of the most common methods 
for the categorization of text and objects. In text classification, the BoW method records the number of 
occurrences of each bag that is created for each instance type or word disregarding the order of the words 
or the grammar. And in visual scene classification it is based on clusters of local descriptors which are 
taken from the images disregarding the order of the clusters. The key idea is to generate a histogram for 
the words in the documents or the features in the images to represent the specified document 
or image. The BoW method is computationally and even conceptually is simpler than many other classification methods. 
For that reason, BoW based systems could record new and higher performance scores on common used benchmarks 
of text and image classification algorithms. This paper presents an overview of BoW, importance of BoW, how does 
it work, applications and challenges of using BoW. This study is useful in terms of introducing the BoW method to 
the new researchers and providing a good background with associated 
related works to the researchers that are working on the model.'''

# Text Preprocessing
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation marks
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize text into words
    tokens = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

# Bag of Words
def bag_of_words(text):
    preprocessed_text = preprocess_text(text)
    # Count word frequencies
    word_counts = Counter(preprocessed_text)
    return word_counts
```

```python
# Apply Bag of Words
bow = bag_of_words(data)
for word,freq in bow.items():
    print(f'{word}-->{freq}')
```

## Performing bag-of-words using CountVectorizer Class

```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample text
data = '''In the past fifteen years, the grow of using Bag of Words (BoW) method in the field of computer 
vision is visibly observed. In addition, for the text classification and texture recognition, it can also 
be used in classification of images, videos, robot localization, etc. It is one of the most common methods 
for the categorization of text and objects. In text classification, the BoW method records the number of 
occurrences of each bag that is created for each instance type or word disregarding the order of the words 
or the grammar. And in visual scene classification it is based on clusters of local descriptors which are 
taken from the images disregarding the order of the clusters. The key idea is to generate a histogram for 
the words in the documents or the features in the images to represent the specified document 
or image. The BoW method is computationally and even conceptually is simpler than many other classification methods. 
For that reason, BoW based systems could record new and higher performance scores on common used benchmarks 
of text and image classification algorithms. This paper presents an overview of BoW, importance of BoW, how does 
it work, applications and challenges of using BoW. This study is useful in terms of introducing the BoW method to 
the new researchers and providing a good background with associated 
related works to the researchers that are working on the model.'''

# Create a CountVectorizer object
vectorizer = CountVectorizer()

# Perform Bag of Words
bow = vectorizer.fit_transform([data])

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Convert sparse matrix to dense array
bow_array = bow.toarray()

# Create dictionary of word frequencies
word_freq = dict(zip(feature_names, bow_array[0]))

print(word_freq)
```

## Performing Tf-Idf using TfidfVectorizer Class

```python
from sklearn.feature_extraction.text import TfidfVectorizer
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation marks
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize text into words
    tokens = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    # Join tokens back into text
    preprocessed_text = ' '.join(filtered_tokens)
    return preprocessed_text

# Preprocess the data
preprocessed_data = preprocess_text(data)
# Create a TfidfVectorizer object
vectorizer = TfidfVectorizer()

# Perform TF-IDF
tfidf = vectorizer.fit_transform([preprocessed_data])

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Convert sparse matrix to dense array
tfidf_array = tfidf.toarray()

# Create dictionary of word TF-IDF scores
word_tfidf = dict(zip(feature_names, tfidf_array[0]))

for word,value in word_tfidf.items():
    print(f'{word} --> {value}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Choose a subset of words to plot
subset_words = feature_names[:50]  # Select first 50 words for plotting

# Get TF-IDF scores for the subset of words
subset_tfidf_scores = tfidf_array[0][:50]  # Take TF-IDF scores for the first document

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(range(len(subset_words)), subset_tfidf_scores, marker='o', color='blue')
plt.title('TF-IDF Scores for Subset of Words')
plt.xlabel('Word Index')
plt.ylabel('TF-IDF Score')
plt.xticks(range(len(subset_words)), subset_words, rotation=90)
plt.tight_layout()
plt.show()
```

## Performing Word2vec to generate the word embeddings

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
import string
```

```python
# Your data corpus
data = '''Artificial intelligence (AI) is transforming various industries, including healthcare, finance, 
  and transportation.
  Machine learning algorithms are used for predictive analytics and decision-making in business 
  applications such as customer relationship management (CRM) and fraud detection.
  Natural language processing (NLP) enables machines to understand and generate human language, 
  facilitating applications like chatbots and language translation.
  Computer vision technology allows computers to interpret and analyze visual information from 
  images and videos, enabling applications like object detection and autonomous vehicles.
  Robotics involves designing, building, and programming robots for tasks such as manufacturing, 
  healthcare assistance, and exploration in hazardous environments.'''

# Preprocessing the data
# Convert text to lowercase
data = data.lower()

# Remove punctuation
data = data.translate(str.maketrans('', '', string.punctuation))

# Tokenize the text
tokens = word_tokenize(data)

# Training the Word2Vec model
# Define Word2Vec model
model = Word2Vec(sentences=[tokens], vector_size=100, window=5, min_count=1, workers=4)

# Train the Word2Vec model
model.train([tokens], total_examples=len(tokens), epochs=10)

# Testing the model
# Get vector for a specific word
vector = model.wv['intelligence']

# Find similar words
similar_words = model.wv.most_similar('intelligence')

# Print results
print("Vector for 'intelligence':", vector)
print("Similar words to 'intelligence':", similar_words)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Extracting vectors and corresponding words
vectors = model.wv[model.wv.key_to_index]
words = list(model.wv.key_to_index.keys())

# Applying PCA to reduce dimensionality to 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(vectors)

# Plotting the scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.5)

# Adding annotations for some words
for i, word in enumerate(words):
    plt.annotate(word, xy=(pca_result[i, 0], pca_result[i, 1]), fontsize=8)

plt.title('Word2Vec Embeddings (2D PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```
